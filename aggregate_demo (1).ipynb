```
results_df = spark.read.parquet (f"{presentation_folder_path}/race_results")
```

```
aggregate_df = results_df.filter("race_year = 2020")
```

```
display(aggregate_df)
```

count , countDistinct , sum

```
from pyspark.sql.functions import count, countDistinct, sum 
```

```
results_df.select(count("race_year")).show()
```

```
aggregate_df.select(count("race_year")).show()
```

```
aggregate_df.select(count("race_name")).show()
```

```
aggregate_df.select(countDistinct("race_name")).show()
```

```
aggregate_df.select(sum("points")).show()
```

```
aggregate_df.where("driver_name = 'Max Verstappen'").select(sum("points"),countDistinct("race_name")) \
.withColumnRenamed("sum(points)", "total_points") \
.withColumnRenamed("count(DISTINCT race_name)", "number_of_races") \
.show()
```

```
aggregate_df.where("driver_name = 'George Russell'").select(sum("points"),countDistinct("race_name")) \
.withColumnRenamed("sum(points)", "total_points") \
.withColumnRenamed("count(DISTINCT race_name)", "number_of_races") \
.show()
```

```
aggregate_df.groupBy("driver_name") \
.agg(sum("points"),countDistinct("race_name")) \
.withColumnRenamed("sum(points)", "total_points") \
.withColumnRenamed("count(race_name)", "number_of_races") \
.show()
```
