Mounting ADLS

```
storage ='formula01dl'
client_id = dbutils.secrets.get(scope='batch2-dbsecret',key ='Batch2-client-id')
tenant_id =dbutils.secrets.get(scope='batch2-dbsecret',key ='batch2-tenant-id')
client_secret =dbutils.secrets.get(scope='batch2-dbsecret',key ='batch2-client-secret')
```

```
config ={
  "fs.azure.account.auth.type": "OAuth",
  "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
  "fs.azure.account.oauth2.client.id": f"{client_id}",
  "fs.azure.account.oauth2.client.secret": f"{client_secret}",
  "fs.azure.account.oauth2.client.endpoint": f"https://login.microsoftonline.com/{tenant_id}/oauth2/token"
}
```

```
def con_mount(container):
  dbutils.fs.mount(source = f"abfss://{container}@{storage}.dfs.core.windows.net/",mount_point = f"/mnt/{storage}/{container}", extra_configs = config)
```

```
con_mount("abhishek")
```

```
%fs
ls dbfs:/mnt/formula01dl/abhishek
```
